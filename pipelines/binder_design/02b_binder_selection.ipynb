### Open and combine input csv(s) into a single dataframe

```python
import pandas as pd
import os
input_dataframe_list = []
for input in inputs:
    if os.path.exists(f"output/{input}"):
        input_dataframe = pd.read_csv(f"output/{input}")
        input_dataframe_list.append(input_dataframe)
combined_csv_dataframe = pd.concat(input_dataframe_list, ignore_index=True)
combined_csv_dataframe
```

### Remove entries without existing pdb files

```python
import os

pdbs = combined_csv_dataframe["model_path"].tolist()
rows_to_remove = []
combined_csv_dataframe = combined_csv_dataframe.reset_index(drop=True) # Resets the dataframe index (when rows are dropped the index of rows isn't updated. For example the first row can have an index of 65)
for pdb in pdbs:
    try:
        if os.path.exists(pdb):
            continue
        else:
            rows_to_remove.append(pdbs.index(pdb))
    except:
        rows_to_remove.append(pdbs.index(pdb))

combined_csv_dataframe = combined_csv_dataframe.drop(rows_to_remove) # Removes entries that fail distance check (or file exists check)
combined_csv_dataframe = combined_csv_dataframe.reset_index(drop=True) # Resets the dataframe index (when rows are dropped the index of rows isn't updated. For example the first row can have an index of 65)
combined_csv_dataframe
```

### Filter binders by binding location on target protein (optional)
Specify the maximum distance from the center of mass of the binder to a target residue on the static protein.

The script, running in a single thread, processes approximately 100 PDB files per minute.

```python
filtered = combined_csv_dataframe
```

```python
import pyrosetta
#from pyrosetta import pose_from_pdb
from pyrosetta import *
import os
pyrosetta.init()


target_residue_index = 134
maximum_distance_from_target = 30

pdbs = filtered["model_path"].tolist()
rows_to_remove = []
filtered = filtered.reset_index(drop=True) # Resets the dataframe index (when rows are dropped the index of rows isn't updated. For example the first row can have an index of 65)
for pdb in pdbs:
    if os.path.exists(pdb):
        pose = pose_from_pdb(pdb)
        # Position of target residue
        target_residue_xyz = pose.pdb_info().pdb2pose("A", target_residue_index)
        # Distance of binder center of mass from a target residue
        binder_distance_from_target_residue = pyrosetta.rosetta.protocols.geometry.centroid_by_chain(pose, 2).distance(pyrosetta.rosetta.protocols.geometry.center_of_mass(pose, target_residue_xyz, target_residue_xyz))
        if binder_distance_from_target_residue > maximum_distance_from_target:
            rows_to_remove.append(pdbs.index(pdb))
        # Optionally add other pyrosetta related filters here
    else:
        rows_to_remove.append(pdbs.index(pdb))
filtered = filtered.drop(rows_to_remove) # Removes entries that fail distance check (or file exists check)

filtered
```

```python
combined_csv_dataframe = filtered
```
