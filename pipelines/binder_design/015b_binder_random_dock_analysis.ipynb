```python
import os
import glob
import subprocess
import pandas as pd
```

# Filter sequences

```python
input = "5fmv_domain4"
all_binders = pd.read_csv(f"output/{input}/random_binders/all_random_binders.csv")
all_binders = all_binders.sort_values("i_pae")
all_binders
```

```python
# How many from each group
top_binders = 1
i_pae_threshold = 7

top_binders_dataframe = (
    all_binders.groupby("name")
    .apply(lambda group: group.nsmallest(top_binders, "i_pae"))
    .reset_index(drop=True)
)
top_binders_dataframe = top_binders_dataframe[
    top_binders_dataframe["i_pae"] <= i_pae_threshold
]
top_binders_dataframe
```

# Prepare folder of best binders

```python
best_random_path = f"output/{input}/random_binders/best_random_binders"
os.makedirs(best_random_path, exist_ok=True)

# Save csv
top_binders_dataframe.to_csv(
    f"output/{input}/random_binders/best_random_binders/best.csv"
)

for i, row in top_binders_dataframe.iterrows():
    pdb = row["model_path"]
    !cp $pdb $best_random_path
```
